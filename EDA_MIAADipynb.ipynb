### Introducci√≥n al EDA
El An√°lisis Exploratorio de Datos (EDA, por sus siglas en ingl√©s) es un paso fundamental en Machine Learning que consiste en examinar, visualizar y resumir los datos antes de aplicar modelos. Su objetivo es comprender la estructura, patrones y anomal√≠as en los datos, lo que ayuda a tomar decisiones informadas sobre la limpieza, transformaci√≥n y selecci√≥n de caracter√≠sticas.


### **Ventajas y Desventajas del EDA**  

### ‚úÖ **Ventajas**  
1. **Identificaci√≥n de patrones y relaciones**  
   - Permite descubrir correlaciones entre variables que pueden mejorar los modelos de Machine Learning.  

2. **Detecci√≥n de valores at√≠picos y errores**  
   - Ayuda a encontrar outliers, datos inconsistentes o valores faltantes que pueden afectar el rendimiento del modelo.  

3. **Comprensi√≥n de la distribuci√≥n de datos**  
   - Permite visualizar c√≥mo est√°n distribuidas las variables (asimetr√≠a, sesgo, varianza).  

4. **Optimizaci√≥n de la selecci√≥n de caracter√≠sticas**  
   - Facilita la eliminaci√≥n de variables irrelevantes o redundantes para mejorar la eficiencia del modelo.  

5. **Mejora en la toma de decisiones**  
   - Un buen EDA ayuda a elegir las mejores t√©cnicas de preprocesamiento y el modelo adecuado para los datos.  

### ‚ùå **Desventajas**  
1. **Puede ser subjetivo**  
   - La interpretaci√≥n de los gr√°ficos y an√°lisis puede depender de la experiencia del analista.  

2. **Toma tiempo y esfuerzo**  
   - En datasets grandes, el EDA puede ser laborioso y consumir recursos computacionales.  

3. **No garantiza buenos modelos**  
   - Aunque ayuda a mejorar la calidad de los datos, el √©xito del modelo depende de otros factores como la selecci√≥n de algoritmos.  

4. **Riesgo de overfitting en preprocesamiento**  
   - Si se ajustan demasiado los datos (por ejemplo, eliminando muchos outliers), el modelo puede perder generalizaci√≥n.  

5. **No siempre es automatizable**  
   - Algunos patrones o anomal√≠as requieren intervenci√≥n manual para ser detectados y corregidos.  

üëâ **Conclusi√≥n:**  
El EDA es un paso crucial en Machine Learning y an√°lisis de datos, ya que permite comprender mejor la estructura de los datos, identificar problemas y tomar decisiones informadas sobre preprocesamiento y modelado. Sin embargo, tambi√©n tiene limitaciones que deben considerarse para evitar sesgos o malinterpretaciones.

# Carga de datos y An√°lisis estad√≠stico b√°sico `.describe()`

## Metodo `.describe()`
El m√©todo `.describe()` en Pandas se emplea en **EDA** y **Machine Learning** para obtener un resumen estad√≠stico de las variables num√©ricas del dataset.  

## **¬øQu√© informaci√≥n proporciona `.describe()`?**  
Cuando se usa `df.describe()`, se generan las siguientes estad√≠sticas clave:  

| Estad√≠stica  | Descripci√≥n |
|-------------|------------|
| **count**   | N√∫mero de valores no nulos en cada columna. |
| **mean**    | Media aritm√©tica de cada variable num√©rica. |
| **std**     | Desviaci√≥n est√°ndar, que mide la dispersi√≥n de los datos. |
| **min**     | Valor m√≠nimo registrado. |
| **25%**     | Primer cuartil (Q1), indica el 25% inferior de los datos. |
| **50%**     | Mediana (Q2), punto central de los datos. |
| **75%**     | Tercer cuartil (Q3), indica el 75% inferior de los datos. |
| **max**     | Valor m√°ximo registrado. |


````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Cargar el dataset
df = sns.load_dataset('diamonds')

# Mostrar las primeras filas del dataset
print(df.head())
print('\n')

#Obtenemos la informacion de las columnas, asi como el tipo de dato que tenemos
print('El tipo de dato de cada columna es: ')
print(df.info())
print('\n')

# Estad√≠sticas descriptivas
print('El EDA de los datos es: ')
print(df.describe())
print('\n')

#Obtenemos la moda de cada una de las columnas(el describe no la ofrece por si sola)
# Obtener la moda de cada columna
moda = df.mode().iloc[0]  # .iloc[0] toma solo el primer valor en caso de m√∫ltiples modas

# Mostrar la moda de cada columna
print("Moda de cada columna:")
print(moda)
print('\n')

# Verificar valores faltantes
print("Valores faltantes:")
print(df.isnull().sum())
````

```Moda (.mode()):```

# üìåEntender la forma de la distribuci√≥n
La distribuci√≥n de una variable num√©rica nos dice c√≥mo est√°n distribuidos los valores de esa variable. Las formas m√°s comunes son:

- Distribuci√≥n normal (campana de Gauss): Los datos est√°n centrados alrededor de la media, con una simetr√≠a en ambos lados.

- Distribuci√≥n sesgada: Los datos est√°n concentrados en un lado y tienen una cola larga en el otro.

  * Sesgo a la derecha (positivo): La cola larga est√° hacia la derecha.
  * Sesgo a la izquierda (negativo): La cola larga est√° hacia la izquierda.
  * Distribuci√≥n multimodal: Los datos tienen m√∫ltiples picos (modas).

  Importancia:
  Saber si una variable tiene una distribuci√≥n normal es crucial para aplicar ciertos m√©todos estad√≠sticos (por ejemplo, pruebas de hip√≥tesis o regresiones lineales), ya que muchos de ellos asumen normalidad.

Si la distribuci√≥n est√° sesgada, puede ser necesario transformar los datos (por ejemplo, aplicar logaritmos) para mejorar el rendimiento de los modelos.

Histograma para los datos y Funcion de densidad de probabilidad

### **¬øQu√© es un Histograma y una PDF en Machine Learning?**  

Tanto los **histogramas** como las **funciones de densidad de probabilidad (PDF)** son herramientas clave para **entender la distribuci√≥n de los datos** en Machine Learning. Veamos en qu√© consisten y c√≥mo se usan. üöÄ  

---

## **1Ô∏è‚É£ ¬øQu√© es un Histograma?**
Un **histograma** es una representaci√≥n gr√°fica que divide los datos en **intervalos (bins)** y muestra **cu√°ntos valores** caen dentro de cada uno.  

üîπ **Se usa para:**  
‚úî Visualizar la **frecuencia** de los datos en un rango.  
‚úî Identificar **sesgo** o **asimetr√≠a** en la distribuci√≥n.  
‚úî Detectar **valores at√≠picos (outliers)**.  

---
## **2Ô∏è‚É£ ¬øQu√© es una PDF (Funci√≥n de Densidad de Probabilidad)?**
La **PDF (Probability Density Function)** describe **c√≥mo se distribuyen los valores de una variable continua** en el espacio de probabilidades.  

üîπ **Diferencias con el histograma:**  
‚úî En lugar de contar ocurrencias, la PDF **suaviza** la distribuci√≥n.  
‚úî La **√°rea total bajo la curva** es igual a **1** (probabilidad total).  
---

## **3Ô∏è‚É£ Comparaci√≥n de Histograma vs PDF**
| üìä **Histograma** | üìà **PDF** |
|------------------|------------|
| Discreto (cuenta valores en intervalos) | Continua (suaviza la distribuci√≥n) |
| Depende del n√∫mero de bins elegidos | No depende de bins |
| Puede mostrar ruido o picos irregulares | Representa mejor la forma real de la distribuci√≥n |

### **üîπ ¬øCu√°ndo usar cada uno?**
‚úî **Histograma** ‚Üí Para explorar los datos de manera r√°pida.  
‚úî **PDF** ‚Üí Para modelar distribuciones y calcular probabilidades.  

---

### **üîπ Conclusi√≥n**
1. El **histograma** muestra la **frecuencia de los datos en intervalos**.  
2. La **PDF** muestra **c√≥mo se distribuyen los valores** en el espacio de probabilidad.  
3. Ambos se usan en **EDA (An√°lisis Exploratorio de Datos)** para entender la estructura de los datos antes de entrenar un modelo de ML.  

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset diamonds de Seaborn
df = sns.load_dataset('diamonds')

# Crear una lista de las columnas del dataset
columns = df.select_dtypes(include=['float64', 'int64']).columns

# Crear los gr√°ficos de PDF e Histograma
plt.figure(figsize=(12, 10))
for i, col in enumerate(columns, 1):
    plt.subplot(3, 3, i)  # Ajustar la cantidad de subgr√°ficos seg√∫n el n√∫mero de columnas
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'PDF y Histograma de {col}', fontsize=12)
    plt.xlabel(col, fontsize=10)
    plt.ylabel('Frecuencia', fontsize=10)

plt.tight_layout()
plt.show()
````

### üåü **Conclusiones del Histograma y PDF ***

- ü™ô **'price' (precio)**: La distribuci√≥n del precio tiene una forma sesgada a la derecha, lo que significa que la mayor√≠a de los diamantes tienen precios m√°s bajos, pero unos pocos tienen precios muy altos. Este sesgo es t√≠pico en distribuciones de precios donde pocos art√≠culos son muy costosos.
  
- üíé **'carat' (peso en quilates)**: La distribuci√≥n de 'carat' es m√°s uniforme y muestra una ligera tendencia a concentrarse en valores m√°s peque√±os, lo que indica que la mayor√≠a de los diamantes son m√°s peque√±os en tama√±o.

- üè∑Ô∏è **'depth'**: Esta columna parece tener una distribuci√≥n casi normal, con la mayor√≠a de los diamantes teniendo un valor intermedio en profundidad, lo que sugiere que las proporciones de los diamantes en este dataset son relativamente constantes.

- üìè **'table'**: Al igual que 'depth', 'table' tiene una distribuci√≥n que se aproxima a una campana de Gauss, lo que indica que las proporciones de la tabla del diamante se distribuyen de manera m√°s o menos uniforme.

- ‚öñÔ∏è **Otras columnas (como 'x', 'y', 'z')**: Estas columnas relacionadas con las dimensiones del diamante muestran distribuciones que tienden a concentrarse en valores m√°s peque√±os, lo que refleja que la mayor√≠a de los diamantes en el dataset son m√°s peque√±os en t√©rminos de sus dimensiones f√≠sicas.

# üìåIdentificar outliers (valores at√≠picos)
Los outliers son valores extremos que se desv√≠an significativamente del resto de los datos. Estos pueden ser errores de medici√≥n, datos incorrectos o valores reales pero inusuales. Estos los vimos en una secion anterior, pero aqui te dejo un resumen de estos elementos

1. Importancia de los Outliers en Machine Learning
  üìâ Impacto en el rendimiento del modelo:
  Los outliers pueden influir fuertemente en el comportamiento de algunos modelos, como la regresi√≥n lineal, donde los valores extremos pueden distorsionar los coeficientes de la ecuaci√≥n. Modelos como los √°rboles de decisi√≥n üå≥ y los modelos basados en vecinos, como k-NN, son menos sensibles a los outliers.

2. üíé Posibilidad de representar casos interesantes:
  No todos los outliers son errores. En muchos casos, los outliers representan casos excepcionales que pueden ser de inter√©s, como productos raros o eventos poco comunes üéØ que podr√≠an tener un valor importante para el an√°lisis.

3. üîç Detecci√≥n de errores de datos:
  Los outliers tambi√©n pueden indicar que algo sali√≥ mal en la recolecci√≥n o entrada de los datos, como errores tipogr√°ficos, mediciones incorrectas o duplicaci√≥n de registros üßπ. Detectarlos puede ayudar a limpiar los datos.

4. ‚öôÔ∏è Estrategia de tratamiento:
  Si un outlier es leg√≠timo, puede ser beneficioso mantenerlo para preservar la integridad del modelo. Si es un error, eliminarlo üóëÔ∏è o transformarlo podr√≠a mejorar el modelo.

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = sns.load_dataset('diamonds')

# Seleccionar solo columnas num√©ricas
numeric_columns = df.select_dtypes(include=[np.number]).columns

# Configurar el tama√±o de la figura
plt.figure(figsize=(15, 10))

# Crear boxplots para cada columna num√©rica
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(2, 4, i)  # 2 filas, 4 columnas de subplots
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot de {col}')
plt.tight_layout()
plt.show()
````

### üåü **Conclusiones de los Boxplots de *Outliers***

1. **üìä Identificaci√≥n de *Outliers***  
   - **Puntos fuera de los bigotes**: Los puntos que se encuentran fuera de los bigotes de la caja son los *outliers*. Estos valores son significativamente m√°s altos o m√°s bajos que el resto de los datos.  
   - **Ejemplo**: Si en la columna **precio** (`price`) hay puntos fuera de los bigotes, eso indicar√≠a que hay diamantes con precios mucho m√°s altos o m√°s bajos üí∞, lo que podr√≠a ser debido a diamantes de caracter√≠sticas muy raras o excepcionales.

2. **üìà Distribuci√≥n de los Datos**  
   - **Forma de la caja**: Si la caja est√° m√°s cerca de uno de los extremos de la escala, podr√≠a indicar que los datos est√°n **sesgados** (asim√©tricos). Esto es √∫til para entender la distribuci√≥n de las variables.  
   - **Ejemplo**: Si la columna **peso en quilates** (`carat`) tiene una caja que se desplaza m√°s hacia el lado derecho ‚û°Ô∏è, podr√≠a indicar que la mayor√≠a de los diamantes son de peso bajo, pero hay algunos diamantes excepcionales muy grandes üíé (sesgo hacia los valores m√°s altos).

3. **üìè Rango Intercuart√≠lico (IQR) y Escala de Valores**  
   - **Tama√±o de la caja**: Una caja **ancha** indica mayor variabilidad entre los cuartiles, lo que puede sugerir que la caracter√≠stica tiene una dispersi√≥n significativa. Una caja **estrecha** sugiere que los datos est√°n concentrados alrededor de la mediana.  
   - **Ejemplo**: En columnas como **claridad** ‚ú® o **corte** üî™, si la caja es peque√±a, puede indicar que la mayor√≠a de los diamantes tienen caracter√≠sticas similares y no hay mucha variabilidad.

4. **üîç Normalidad de los Datos**  
   - **Simetr√≠a de los bigotes**: Si los bigotes son casi iguales en longitud a ambos lados de la caja, eso sugiere que los datos est√°n relativamente **sim√©tricos**. Si un bigote es m√°s largo que el otro, eso sugiere que los datos est√°n **sesgados** hacia uno de los extremos.  
   - **Ejemplo**: Si en la columna **precio** se ve un bigote m√°s largo en el lado superior üîù, esto indicar√≠a que hay m√°s diamantes con precios significativamente altos üíéüíµ, lo que puede reflejar una tendencia en el mercado o la presencia de diamantes de alta gama.

5. **‚ö†Ô∏è Posibles Errores en los Datos**  
   - **Puntos aislados que no parecen tener sentido**: Si los *outliers* son puntos que claramente no tienen sentido ‚ùå (por ejemplo, un precio de $0 o un peso en quilates negativo), podr√≠an ser **errores de datos**. En este caso, es importante investigar estos valores para ver si se deben corregir o eliminar üóëÔ∏è.

6. **üíé Valores T√≠picos vs. Valores Raros**  
   - **Diamantes raros o excepcionales**: Los *outliers* pueden representar casos raros o excepcionales, como diamantes de gran tama√±o o de calidad especial üåü, que son relevantes para ciertas aplicaciones o an√°lisis de mercado üìä. Estos valores pueden ser importantes si se desea comprender los segmentos de alta gama del mercado de diamantes.

# üî• **Criterios Utilizados en el Heatmap**  

Un **heatmap de correlaci√≥n** se basa en la **matriz de correlaci√≥n**, que mide la relaci√≥n entre variables num√©ricas.  

## üìå **Principales Criterios Usados en el Heatmap**  

### **1Ô∏è‚É£ M√©todo de Correlaci√≥n**  
Los heatmaps suelen calcular correlaciones mediante:  

üìå **Coeficiente de Pearson (üî¥ m√°s com√∫n en ML)**  
- Mide la **relaci√≥n lineal** entre dos variables.  
- Va de **-1** (correlaci√≥n negativa perfecta) a **1** (correlaci√≥n positiva perfecta).  
- Si es **cercano a 0**, indica que no hay relaci√≥n lineal.  

üìå **Coeficiente de Spearman**  
- Se usa si la relaci√≥n es **mon√≥tona pero no lineal**.  
- Se basa en rangos en lugar de valores reales.  

üìå **Coeficiente de Kendall**  
- Similar a Spearman pero con un enfoque en dependencias ordinales.  

‚úÖ En EDA y ML, generalmente usamos **Pearson**, a menos que haya relaciones no lineales.  

---

### **2Ô∏è‚É£ Selecci√≥n de Variables Num√©ricas**  
- Solo se incluyen **columnas num√©ricas**, ya que las categ√≥ricas no tienen sentido en una matriz de correlaci√≥n.  
- Si hay valores **categ√≥ricos codificados como n√∫meros**, es recomendable tratarlos aparte.

- Se definen **l√≠neas divisorias (`linewidths=0.5`)** para mejorar la visualizaci√≥n.

# üìå **An√°lisis de Normalidad en EDA y ML**  

El **an√°lisis de normalidad** eval√∫a si una variable sigue una **distribuci√≥n normal (Gaussiana)**. Es clave en **EDA** y **ML** porque muchas t√©cnicas estad√≠sticas asumen normalidad en los datos.  

---

## üéØ **¬øPor qu√© es Importante en EDA y ML?**  

### ‚úÖ **1. Influencia en Modelos Predictivos**  
üîπ Modelos como la **Regresi√≥n Lineal** y **ANOVA** asumen normalidad en los datos para hacer inferencias v√°lidas.  
üîπ Si los datos no son normales, podr√≠amos necesitar transformaciones (**log, Box-Cox**).  

### ‚úÖ **2. Impacto en Pruebas Estad√≠sticas**  
üìä Muchas pruebas estad√≠sticas, como **t-test** o **intervalos de confianza**, requieren que los datos sean normales para que los resultados sean confiables.  

### ‚úÖ **3. Manejo de Valores Extremos (Outliers)**  
üìä Si los datos tienen una **distribuci√≥n sesgada**, es m√°s probable que existan **valores at√≠picos**, lo que puede afectar el desempe√±o del modelo.  

### ‚úÖ **4. Selecci√≥n de Algoritmos de Machine Learning**  
üìä Algunos modelos como **Naive Bayes** y **SVM con kernel RBF** funcionan mejor cuando los datos se distribuyen normalmente.  

---

## üî• **¬øC√≥mo Evaluar la Normalidad?**  

### **üìä M√©todos Gr√°ficos**  
üìä **Histogramas** ‚Üí Para observar la forma de la distribuci√≥n.  
üìä **QQ-Plot (Quantile-Quantile Plot)** ‚Üí Compara cuantiles de los datos con los de una normal te√≥rica.  
üìä **Boxplot** ‚Üí Detecta asimetr√≠as y valores at√≠picos.  

### **üìä Pruebas Estad√≠sticas**  
üìä **Shapiro-Wilk Test** (Para datasets peque√±os, \( n < 5000 \))  
üìä **Kolmogorov-Smirnov Test** (Para muestras grandes)  
üìä **Anderson-Darling Test** (M√°s estricto que Shapiro-Wilk)  
üìä **D'Agostino K¬≤ Test** (Mide asimetr√≠a y curtosis)  



````python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np

# Cargar dataset
df = sns.load_dataset('diamonds')

# Seleccionar solo columnas num√©ricas
numeric_columns = df.select_dtypes(include=[np.number]).columns

# Funci√≥n para realizar pruebas de normalidad
def normality_tests(data, column_name):
    print(f"\nPruebas de normalidad para la columna: {column_name}")

    # Prueba de Shapiro-Wilk
    shapiro_test = stats.shapiro(data.sample(min(5000, len(data)), random_state=42))  # Muestra de 5000 para evitar problemas de c√≥mputo
    print(f"Shapiro-Wilk Test: p-valor = {shapiro_test.pvalue:.5f}")

    # Prueba de Kolmogorov-Smirnov
    ks_test = stats.kstest(data, 'norm', args=(data.mean(), data.std()))
    print(f"Kolmogorov-Smirnov Test: p-valor = {ks_test.pvalue:.5f}")

    # Prueba de Anderson-Darling
    anderson_test = stats.anderson(data, dist='norm')
    print(f"Anderson-Darling Test: Estad√≠stico = {anderson_test.statistic:.5f}")
    for i in range(len(anderson_test.critical_values)):
        sl, cv = anderson_test.significance_level[i], anderson_test.critical_values[i]
        if anderson_test.statistic < cv:
            print(f"  Al nivel de significancia {sl}%, los datos parecen normales.")
        else:
            print(f"  Al nivel de significancia {sl}%, los datos NO parecen normales.")

    # Prueba de D'Agostino-Pearson
    dagostino_test = stats.normaltest(data)
    print(f"D'Agostino-Pearson Test: p-valor = {dagostino_test.pvalue:.5f}")

# Funci√≥n para graficar histograma y QQ-Plot
def plot_distribution(data, column_name):
    plt.figure(figsize=(12, 5))

    # Histograma
    plt.subplot(1, 2, 1)
    sns.histplot(data, bins=50, kde=True, color="skyblue")
    plt.title(f"Distribuci√≥n de {column_name}")

    # QQ-Plot
    plt.subplot(1, 2, 2)
    stats.probplot(data, dist="norm", plot=plt)
    plt.title(f"QQ-Plot de {column_name}")

    plt.show()

# Aplicar pruebas de normalidad y gr√°ficos a cada columna num√©rica
for column in numeric_columns:
    data = df[column]
    plot_distribution(data, column)
    normality_tests(data, column)
````

### üåü **Conclusiones del An√°lisis de Normalidad**
Las pruebas de normalidad nos ayudan a determinar si los datos siguen una **distribuci√≥n normal** (tambi√©n conocida como distribuci√≥n gaussiana). Una distribuci√≥n normal es importante porque muchos m√©todos estad√≠sticos y modelos de machine learning asumen que los datos son normales.

- **Si los datos son normales**:
  - Podemos usar m√©todos param√©tricos (por ejemplo, pruebas t, ANOVA, regresi√≥n lineal).
  - Los intervalos de confianza y los valores p son m√°s precisos.
